## Создадим единый DataFrame с необходимыми признаками

- AGREEMENT_RK — уникальный идентификатор объекта в выборке;
- TARGET — целевая переменная: отклик на маркетинговую кампанию (1 — отклик был зарегистрирован, 0 — отклика не было);
- AGE — возраст клиента;
- SOCSTATUS_WORK_FL — социальный статус клиента относительно работы (1 — работает, 0 — не работает);
- SOCSTATUS_PENS_FL — социальный статус клиента относительно пенсии (1 — пенсионер, 0 — не пенсионер);
- GENDER — пол клиента (1 — мужчина, 0 — женщина);
- CHILD_TOTAL — количество детей клиента;
- DEPENDANTS — количество иждивенцев клиента;
- PERSONAL_INCOME — личный доход клиента (в рублях);
- LOAN_NUM_TOTAL — количество ссуд клиента;
- LOAN_NUM_CLOSED — количество погашенных ссуд клиента.

Признаки AGE, SOCSTATUS_WORK_FL, SOCSTATUS_PENS_FL, GENDER, CHILD_TOTAL и DEPENDANTS возьмем из датафрейма d_clients, соответствующего файлу "D_clients.csv". 
Изменим название столбца "ID" на "ID_CLIENT", чтобы оно совпадало с аналогичными столбцами в других датафреймах.

```python
data = pd.DataFrame(d_clients.sort_values(by='ID')[['ID', 'SOCSTATUS_WORK_FL', 'SOCSTATUS_PENS_FL', 'GENDER', 'CHILD_TOTAL',
                                                    'DEPENDANTS']]).reset_index(drop=True)
data.columns = ['ID_CLIENT'] + list(data.columns[1:])
data
```

Размер d_clients больше, чем размер d_target. Следовательно, в d_clients содержится информация о клиентах, для которых не известно значение целевой переменной, хранящееся в d_target. Поэтому, при дальнейшем объединении датафреймов, будут учитываться только ID клиентов, содержащиеся в d_target.

Добавим признаки AGREEMENT_RK и TARGET в новый итоговый датафрейм.

```python
data = data.merge(d_target.sort_values(by='ID_CLIENT'), on='ID_CLIENT', how='inner')
data
```

Добавим признак PERSONAL_INCOME в новый итоговый датафрейм.

```python
tmp = d_salary.groupby('ID_CLIENT').sum().reset_index().sort_values(by='ID_CLIENT')['PERSONAL_INCOME']
data.insert(len(data.columns), 'PERSONAL_INCOME', tmp)
data
```

Добавим признаки LOAN_NUM_TOTAL и LOAN_NUM_CLOSED в новый итоговый датафрейм.

```python
tmp = d_loan.merge(d_close_loan, on='ID_LOAN').groupby('ID_CLIENT').count().reset_index()
data.insert(len(data.columns), 'LOAN_NUM_TOTAL', tmp['ID_LOAN'])
tmp = d_loan.merge(d_close_loan, on='ID_LOAN').groupby('ID_CLIENT').sum().reset_index()
data.insert(len(data.columns), 'LOAN_NUM_CLOSED', tmp['CLOSED_FL'])
data
```

Отбросим параметр ID_CLIENT из итогового датафрейма.

```python
data.drop(['ID_CLIENT'], inplace=True, axis=1)
```

### Собран датафрейм, содержащий все необходимые признаки, указанные в задании.
